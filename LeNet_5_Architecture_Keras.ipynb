```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_lenet5(input_shape=(32, 32, 1), num_classes=10):
    model = models.Sequential()

    # Layer C1: Convolutional
    model.add(layers.Conv2D(filters=6, kernel_size=(5, 5), strides=(1, 1),
                            activation='tanh', input_shape=input_shape))

    # Layer S2: Average Pooling
    model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)))

    # Layer C3: Convolutional
    model.add(layers.Conv2D(filters=16, kernel_size=(5, 5), strides=(1, 1),
                            activation='tanh'))

    # Layer S4: Average Pooling
    model.add(layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2)))

    # Layer C5: Convolutional (Outputs 1x1x120 given a 32x32 input image)
    model.add(layers.Conv2D(filters=120, kernel_size=(5, 5), strides=(1, 1),
                            activation='tanh'))

    # Flatten the output for Dense layers
    model.add(layers.Flatten())

    # Layer F6: Fully Connected
    model.add(layers.Dense(units=84, activation='tanh'))

    # Output Layer
    # Using softmax here, which is the standard for modern classification tasks
    model.add(layers.Dense(units=num_classes, activation='softmax'))

    return model

# Example usage:
# model = build_lenet5()
# model.summary()
```
