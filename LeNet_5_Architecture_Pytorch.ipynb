```python
import torch
import torch.nn as nn

class LeNet5(nn.Module):
    def __init__(self, num_classes=10):
        super(LeNet5, self).__init__()

        # Feature Extractor
        self.features = nn.Sequential(
            # Layer C1: Convolutional (1 input channel, 6 output channels, 5x5 kernel)
            nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1),
            nn.Tanh(),
            # Layer S2: Average Pooling (2x2 kernel, stride 2)
            nn.AvgPool2d(kernel_size=2, stride=2),

            # Layer C3: Convolutional (6 input channels, 16 output channels, 5x5 kernel)
            nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1),
            nn.Tanh(),
            # Layer S4: Average Pooling (2x2 kernel, stride 2)
            nn.AvgPool2d(kernel_size=2, stride=2),

            # Layer C5: Convolutional (Acts as a fully connected layer since input is 5x5)
            nn.Conv2d(in_channels=16, out_channels=120, kernel_size=5, stride=1),
            nn.Tanh()
        )

        # Classifier
        self.classifier = nn.Sequential(
            # Layer F6: Fully Connected (120 inputs, 84 outputs)
            nn.Linear(in_features=120, out_features=84),
            nn.Tanh(),
            # Output Layer: Fully Connected (84 inputs, 10 outputs for classes)
            nn.Linear(in_features=84, out_features=num_classes)
        )

    def forward(self, x):
        # Pass through feature extractor
        x = self.features(x)
        # Flatten the output for the linear layers
        x = torch.flatten(x, 1)
        # Pass through classifier
        x = self.classifier(x)
        return x

# Example usage:
# model = LeNet5()
# dummy_input = torch.randn(1, 1, 32, 32) # (batch_size, channels, height, width)
# output = model(dummy_input)
```
